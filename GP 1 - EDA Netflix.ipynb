Вследствие этого, крайне важно ответственно подойти к выбору сокомандников, распределению обязанностей внутри команды, а также пресечению наличия "мертвых душ" в команде.

При необходимости, команда большинством голосов имеет право исключить человека из команды *до защиты проекта*, сообщив об этом преподавателю, в случае если такой человек никак не проявлялся в ходе выполнения проекта и являлся как раз таки той самой "мертвой душой". Помните про эту возможность, поскольку наличие "мертвых душ" может негативно повлиять на качество и оценку для всего проекта и, как следствие, в целом для всей команды!

## Оценивание и штрафы



Групповой проект оценивается исходя из устной защиты (презентации) командой выполненного ими проекта перед комиссией. Комиссия составляется из преподавателей курса, ассистентов курса, а также приглашенных экспертов из индустрии.

В рамках каждой защиты комиссия выставляет два типа оценок.

Первая оценка — это оценка для всего проекта в целом. Она ставится комиссией, исходя из глубины проработки командой предметной области, сложности применяемых в проекте инструментов, а также прочих аспектов, связанных с содержанием выполняемого в проекте ТЗ. При выставлении этой оценки комиссия ориентируется на критерии, более подробно представленные ниже в разделе "Критерии оценки проекта".

Вторая оценка — это индивидуальная оценка каждого участника команды. Именно эта оценка выставляется впоследствии в ведомость. Индивидуальная оценка участника команды определяется комиссией, исходя из того, насколько участник ориентируется в выполненном командой проекте, как он отвечает на адресованные персонально ему вопросы комиссии, насколько в принципе прослеживается его роль в проекте, а также исходя из других аспектов, связанных с участием конкретного человека в деятельности команды. Индивидуальная оценка не может превосходить оценку команды за проект в целом (первую оценку), однако, в случае когда участник спокойно и корректно отвечает на все адресованные ему вопросы, свободно ориентируется в материале и содержании выполненного проекта, а также когда его роль в работе команды не вызывает сомнений, — в таком случае, индивидуальная оценка участника становится равной оценке для всего проекта (т.е. в идеальном случае — вторая оценка равна первой оценке).

Во избежание субъективности и/или предвзятости оценки тех или иных членов комиссии в отношении тех или иных команд, а также каких-либо иных проявлений неравных условий — распределение команд между комиссиями осуществляется абсолютно рандомно с использованием средств библиотеки Numpy :)) — в качестве рэндом сида выступает дата, в которую производится распределение, например, `np.random.seed(18092025)`.

Обратите, пожалуйста, внимание на то, что бОльшая часть комиссий осуществляет приём проектов в очной форме, однако могут присутствовать и 1-2 комиссии, которые будут работать дистанционно с использованием средств видеосвязи. В случае попадания команды путем случайного распределения в комиссию, работающую в очной форме, *не допускается* какое-либо изменение формата проведения для команды на онлайн-режим, равно как и любые другие перемещения команд между комиссиями (смены комиссий), для сохранения честности и непредвзятости случайного распределения.

Явка студента на защиту проекта является обязательной; вклад студента без его присутствия не оценивается и автоматически является равным нулю. Для дистанционных комиссий, дополнительным требованием является обязательное наличие включенной камеры в течение всего времени нахождения в конференции

#### Удалите из данных дубликаты.

- Много ли их? В каких группах их больше всего?

```python
data_without_dublicates = data.drop_duplicates()
data_without_dublicates = data_without_dublicates.reset_index(drop=True) # всего 500 строк-дубликатов
dublicates = data[data.duplicated()]
dublicates = dublicates.reset_index(drop=True)

count_rate_dubl = dublicates.groupby(['rating']).count()
count_rate_data = data_without_dublicates.groupby('rating').count()

sorted_title = count_rate_dubl.sort_values(['title'], ascending=False)
mx_groups = sorted_title.iloc[0:3, 0] # Топ-3 группы с наибольшим числом дубликатов
mx_groups
```

#### Сколько рейтинговых групп представлено в данных?

```python
count_rate_data_without_dubl = data_without_dublicates.groupby(['rating']).count()
rate_groups_count = len(count_rate_data_without_dubl)
print('Количество уникальных групп из датасета без дубликатов:', rate_groups_count)
rate_data_without_dub = data_without_dublicates['rating'].unique()
rate_data_without_dub = rate_data_without_dub.tolist()
print('Уникальные типы рейтинговых групп:', end = ' ')
print(*rate_data_without_dub, sep = ', ')
```

#### Какие оценки пользователи ставят чаще?

- Постройте гистограмму оценок.
- Почему именно такие оценки ставят чаще всего?

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data_without_dublicates['user rating score'] = data_without_dublicates['user rating score'].ffill()

plt.figure(figsize=(20, 12))
gistogram = sns.histplot(data_without_dublicates['user rating score'], bins = 50, color = 'green', alpha=0.6)
plt.title('Distribution of the number of people by URS', fontsize = '18')
plt.xlabel('User rating score (URS)', fontsize = '14')
plt.ylabel('Number of films', fontsize = '14')
plt.xticks(np.arange(55, 101, 1))
plt.yticks(np.arange(0, 30, 5))

plt.show()


```

```python
import plotly.graph_objects as go

mean_for_rating = data_without_dublicates.groupby(['rating'])['user rating score'].mean()
order_for_rating = sorted(data_without_dublicates['rating'].unique())
mean_for_rating = mean_for_rating.reindex(order_for_rating)

scatter = go.Scatter(
    x=data_without_dublicates['rating'],
    y=data_without_dublicates['user rating score'],
    mode='markers',
    name='User rating score',
    marker=dict(color='#8172B2', opacity=0.5)
)

line = go.Scatter(
    x=mean_for_rating.index,
    y=mean_for_rating.values,
    mode='lines+markers',
    name='Mean rating',
    line=dict(color='#55A868', width=3),
    marker=dict(symbol='diamond')
)

layout = go.Layout(
    title=dict(text='Distribution of ratings and film categories', font=dict(size=16), x=0.5),
    xaxis=dict(title='Types of shows', categoryorder='array', categoryarray=order_for_rating),
    yaxis=dict(title='User rating score', tickvals=np.arange(50, 101, 5)),
    showlegend=True,
    font=dict(size=12),
    width=1100,
    height=800
)

fig = go.Figure(data=[scatter, line], layout=layout)
fig.show()
```

#### Выведите основную информацию об оценках пользователей
- Чему равны среднее, стандартное отклонение, минимум, максимум, медиана?
- Отличаются ли медиана и среднее?
- Могут ли данные характеристики значительно отличаться? Почему?

```python
avg_value = data_without_dublicates['user rating score'].mean()
std_value = data_without_dublicates['user rating score'].std() 
mn_value = data_without_dublicates['user rating score'].min()
mx_value = data_without_dublicates['user rating score'].max()
mdn_value = data_without_dublicates['user rating score'].median()

print('Среднее значение:', round(avg_value, 2))
print('Стандартное отклонение:', round(std_value, 1))
print('Минмум:', mn_value)
print('Максимум:', mx_value)
print('Медиана:', mdn_value)
```

#### Теперь вам нужно выбрать любое из представленных шоу и при помощи изученных инструментов составить описательный портрет этого шоу :)

```python
# Импортируем 1 датасет с выручкой Netflix по годам
turnover = pd.read_csv('Turnover-Netflix.csv')
turnover = turnover.drop(columns=['Number of Subscriptions', 'Price per share in US', 'Employees'])

# Импортируем 2 датасет с датами добавления всех шоу на платформу Netflix (целевая хар-ка - release on Netflix)
data_with_release_on_netflix = pd.read_csv('netflix_release.csv')
data_with_release_on_netflix['release on Netflix'] = data_with_release_on_netflix['release on Netflix'].astype(int)

# Импортируем 3 датасет с процентным соотношением сборов каждого рейтинга в каждом году
revenue_by_rating = pd.read_csv('rev_by_rating.csv')
general = pd.merge(data_with_release_on_netflix, turnover, left_on='release on Netflix', right_on='Year', how='inner').drop(columns=['Year', 'Balance sheet profit in million'])

# Считаем кол-во фильмов в каждом рейтинге за год и кол-во фильмов в году
count_rating_per_year = general.groupby(['release on Netflix', 'rating']).size().reset_index(name='count_rating_per_year')
count_per_year = general.groupby('release on Netflix')['title'].count().reset_index(name='count_per_year')
joined1 = general.merge(count_rating_per_year, on=['release on Netflix', 'rating'], how='left')
joined2 = pd.merge(joined1, count_per_year, on='release on Netflix', how='left')

np.random.seed(241241)
joined3 = pd.merge(joined2, revenue_by_rating, left_on=['release on Netflix', 'rating'], right_on=['year', 'rating'], how='inner').drop(columns=['year', 'release year'])
joined3['avg_revenue_by_film'] = (joined3['Turnover in million'] * joined3['percentage_by_rating'] / joined3['count_rating_per_year']) * np.random.uniform(0.85, 1.05, size=len(joined3))
revenue = joined3.groupby(['release on Netflix'])['avg_revenue_by_film'].sum()
joined3 = joined3.drop(columns=['count_rating_per_year', 'count_per_year'])

# Рассматриваем ТОЛЬКО до 2017 года, как в исходном датасете
joined3 = joined3[joined3['release on Netflix'] <= 2017]
joined3
```
